```python
#Turning pandas variables into list variables
data1 = pandas_data1.values.tolist()
data2 = pandas_data2.values.tolist()
```

###Analysis of Histograms
Data2 (the Pvalb data) seems to have a lower resting membrane potential compared to data1 (the SST data). The interspike interval time (ISI) of data1 appears faster compared to the histogram showing the ISI of data2. A shorter ISI suggests that the resting membrane potential is closer to the threshold for an action potential, because less time is needed between action potentials, and thus shorter ISI.
