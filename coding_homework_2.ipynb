```python
#importing numpy, plt, and statistics
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
#Importing data (v_rest_sst and v_rest_pvalb) as numpy arrays
##importing sst data and setting to v_rest1 variable
sst_url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest1 = np.loadtxt(sst_url)
##importing pvalb data and setting to v_rest2 variable
pvalb_url = sst_url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest2 = np.loadtxt(pvalb_url)
```

```python
#concatenate v_rest_sst and v_rest_pvalb into a matrix
row1 = np.array(v_rest1)
row2 = np.array(v_rest2)
v_rest = np.vstack ((row1,row2))
#checking v_rest matrix shape
print(v_rest.shape)
```

```python
#size of v_rest matrix
print(v_rest.size)
```

```python
#printing first row of v_rest
print(v_rest[0])
```

```python
#checking if returning correct # of items in first row of v_rest
v_rest_check = (v_rest[0])
print(v_rest_check.size)
```

```python
#finding sum of third column of v_rest
v_rest[:,2].sum()
```

```python
#print second to last value in second row of v_rest
v_rest[1, -2]
```

```python
#create histogram of first row of values of v_rest
plt.hist(v_rest[0])

#label axes
plt.ylabel('Counts')
plt.xlabel('Resting membrane potential (mV)')
```

```python
#mean of first row of v_rest
v_rest_mean_sst = st.mean(v_rest[0])
print(v_rest_mean_sst)

#mean of second row of v_rest
v_rest_mean_pvalb = st.mean(v_rest[1])
print(v_rest_mean_pvalb)
```

The means of the sst row is bigger than the mean of the second row, the pvalb row.

```python
#import scipy.stats
import scipy.stats as stats

```

```python
#t-test on two rows of v_rest
stats.ttest_ind(v_rest[0],v_rest[1])
```

The p-value is much smaller than 0.05 (p = 7.159828660199036e-09), so there is a statistically significant difference between the two rows, and therefore between the sst and the pvalb means.
