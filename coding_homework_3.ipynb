```python
#import numpy plots and stats
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import scipy.stats as stats
```

```python
#import data
v_rest_sst = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv')
v_rest_pvalb = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv')
```

```python

#mean, t-test, hist, comparing pvalb and sst
v_rest_mean_sst = st.mean(v_rest_sst)
print(v_rest_mean_sst)
v_rest_mean_pvalb = st.mean(v_rest_pvalb)
print(v_rest_mean_pvalb)


#t-test

t_stats, p_value = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
print(t_stats)
print(p_value)

#histogram
if p_value > 0.05:
  fig, ax = plt.subplots(1,2)
  ax[0].hist(v_rest_sst)
  ax[0].set_title("v_rest_sst")
  ax[0].set_xlabel("ISI")
  ax[0].set_ylabel("Count")
  ax[1].hist(v_rest_pvalb)
  ax[1].set_title("v_rest_pvalb")
  ax[1].set_xlabel("ISI")
  ax[1].set_ylabel("Count")

  plt.show


```

```python
#creating vrestHist
def vrestHist():
  _rest_mean_sst = st.mean(v_rest_sst)
  print(v_rest_mean_sst)
  v_rest_mean_pvalb = st.mean(v_rest_pvalb)
  print(v_rest_mean_pvalb)
  t_stats, p_value = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
  print(t_stats)
  print(p_value)
  if p_value > 0.05:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(v_rest_sst)
    ax[0].set_title("v_rest_sst")
    ax[0].set_xlabel("ISI")
    ax[0].set_ylabel("Count")
    ax[1].hist(v_rest_pvalb)
    ax[1].set_title("v_rest_pvalb")
    ax[1].set_xlabel("ISI")
    ax[1].set_ylabel("Count")
    plt.show
```

```python
#calling vrestHist
vrestHist()
```
