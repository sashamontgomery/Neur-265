###<b>Task 1:</b>

1. Import the <code>patch_seq_expanded_v2</code> .csv file from our GitHub repository as a pandas dataframe called <code>patch_seq_v2</code>.
2. Create a function that:
    - Makes three subplots contained in one row (one row, three columns)
    - Loops through each row of your <code>patch_seq_v2</code> variable
    - Plots 'ISI' vs. 'Fast_Trough' as a scatter plot in the first (leftmost) subplot if 'Marker' is 'Pvalb', plots 'ISI' vs. 'Fast_Trough' in the second (middle) subplot if 'Marker' is 'Sst', and plots 'ISI' vs. 'Fast_Trough' in the third (rightmost) subplot if 'Marker' is 'Vip'.
    - For each subplot, dots should be red if 'Fast_Trough' is less than -60 *and* 'ISI' is less than 60. If these conditions are not met, dots should be blue.
    - Labels axes on all three subplots
3. Make a new code cell that calls your function. Your function should produce the three subplots when called.


```python
#import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import csv and turn into pandas dataframe

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/patch_seq_expanded_v2.csv'
patch_seq_v2 = pd.read_csv(url)
#viewing patch seq
print(patch_seq_v2)
```

```python
#make three subplots contained in one row
patch_length = len(patch_seq_v2)
fig,ax = plt.subplots(1,3,figsize=(15,5))
for i in range(patch_length):
  x = patch_seq_v2['ISI'].loc[i]
  y = patch_seq_v2['Fast_Trough'].iloc[i]
  marker = patch_seq_v2['Marker'].iloc[i]
  ax[0].set_ylabel('Fast Trough')
  ax[0].set_xlabel('ISI')
  ax[1].set_ylabel('Fast Trough')
  ax[1].set_xlabel('ISI')
  ax[2].set_ylabel('Fast Trough')
  ax[2].set_xlabel('ISI')
  if marker == 'Pvalb':
      if y[i] < -60 and x[i] < 60:
        ax[0].scatter(x[i], y[i], color='red')
      elif y[i] < -60 and x[i] < 60:
        ax[0].scatter(x[i], y[i], color='blue')
  if marker == 'Sst':
      if y[i] < -60 and x[i] < 60:
        ax[1].scatter(y[i], x[i], color='red')
      if y[i] < -60 and x[i] < 60:
        ax[1].scatter(x[i], y[i], color='blue')
  if marker == 'Vip':
      if y[i] < -60 and x[i] < 60:
        ax[2].scatter(x[i], y[i], color='red')
      elif y[i] < -60 and x[i] < 60:
        ax[2].scatter(x[i], y[i], color='blue')
```
