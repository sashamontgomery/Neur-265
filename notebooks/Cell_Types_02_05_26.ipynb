### <i> Patch-Seq summary <i>

Patch-seq allowed for the researchers to characterize the transcriptomes and physiological properties of neurons (specifically of mouse visual cortical GABAergic intervals).
Data was obtained via patch clamping, where sections of the neuron were clamped and then electrical signaling data was analyzed. After the patch clamp, the neuron was dye filled to identify the morphology and then the RNA was sequenced to evaluate the gene sequence.

Some advantages include being able to directly analyze the activity of the neurons and electrical patterns from single neurons. The activity of single neurons can be evaluated, not only larger groups or areas of neurons. This also has the advantage of being able to look at gene expression, gene morphology, and RNA sequencing.
