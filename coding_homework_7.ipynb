```python
#importing relevant modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading data
acc_hits = np.loadtxt("https://raw.githubusercontent.com/sashamontgomery/Neur-265/refs/heads/main/acc_hits_hw5.csv", delimiter = ",")
acc_fa = np.loadtxt("https://raw.githubusercontent.com/sashamontgomery/Neur-265/refs/heads/main/acc_fa_hw5.csv", delimiter = ",")
```

```python
#printing shape of each array
print("acc_hits shape:", acc_hits.shape)
print("acc_fa shape:", acc_fa.shape)
```

```python
#creating time variable
time = np.linspace(0, 4, 8001)
```

```python
# Finding sampling rate
sampling_rate = 8001/4
print(sampling_rate)
```

```python
#plotting first trials of acc_hits and acc_fa
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(time, acc_hits[0,:], color = 'blue')
ax[1].plot(time, acc_fa[0,:], color = 'red')
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Voltage")
ax[0].set_title("acc_hits")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Voltage")
ax[1].set_title("acc_fa")


```

I think the dominant oscillation in the LFP for the hits is 3z and for the false alarms is 1 hz.

```python
#plotting ERP of acc_hits
plt.plot(time, acc_hits.mean(0))        # Plot the ERP of acc_hits
plt.xlabel('Time [s]')           # Label the axes
plt.ylabel('Voltage')
plt.title('ERP of acc_hits')  # ... provide a title
plt.hlines(0, time[0], time[-1], color = "black")
plt.vlines(2, -50, 50, 'k', lw=2)
```

Do you see any evidence of consistent acitvity in the LFP surrounding screen touch (vertical black bar)?

Yes, after two seconds, I can see there are increases in voltage, suggesting consistency across trials because the voltages are not canceling eachother out to appear more flat.

```python
#plotting acc_hits and acc_fa
plt.plot(time, acc_hits.mean(0))        # Plot the ERP of acc_hits
plt.xlabel('Time [s]')           # Label the axes
plt.ylabel('Voltage')
plt.plot(time, acc_fa.mean(0), color = "blue")        # Plot the ERP of acc_fa
plt.title('ERP of acc_hits and acc_fa')  # ... provide a title
plt.hlines(0, time[0], time[-1], color = "black")
plt.vlines(2, -50, 50, 'k', lw=2)
```

Are there any differences in the ERPs between hits and false alarms?  

The ERPs for the false alarms are flatter, meaning they are canceling eachother out across different trials. On the other hand, the hits are less flat, showing more consistency, especially after 2 seconds.
